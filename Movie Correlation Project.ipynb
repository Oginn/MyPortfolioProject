```python
# Importing Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure


%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjsusting the configuration of the plot creating.
```

```python
# Reading in the dataset

df = pd.read_csv(r'C:\Users\OMATOLOGY-TECH\Downloads\movies.csv')
```

```python
df.shape
```

```python
#A real Quick look at the head data

df.head()
```

```python
# Replace missing values in "budget" with the median value
budget_median = df['budget'].median()
gross_median = df['gross'].median()
score_median = df['score'].median()
votes_median = df['votes'].median()
runtime_median = df['runtime'].median()

df['budget'].fillna(budget_median, inplace=True)
df['gross'].fillna(gross_median, inplace=True)
df['score'].fillna(score_median, inplace=True)
df['votes'].fillna(votes_median, inplace=True)
df['runtime'].fillna(runtime_median, inplace=True)
```

```python
# Cleaning up the dataset
# Checking for missing data

for col in df.columns:
    missing_per = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_per))
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
df.head()
```

```python
#Checking the data type
df.dtypes
```

```python
#Change the data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Displaying all the dataset

pd.set_option('display.max_row', None)

df = df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python

# Drop any duplicates ( to display all the unique values)
# Converting to str data type
df['company'] = df['company'].astype(str)


df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scarter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Gross Earning')
plt.ylabel('Budget for Film Production')



plt.show()
```

```python
# Plot Budget Vs Gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "yellow"}, line_kws={"color": "green"})
```

```python
# Correlation
# Compute correlation matrix
corr_matrix = df.corr()

# Print correlation matrix
print(corr_matrix)
```

```python
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Feactures, \n')

plt.xlabel('Movie Feactures')
plt.ylabel('Movie Feactures')

plt.show()
```

```python
# Looking at Company
df_numerized = df

for  col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Feactures, \n')

plt.xlabel('Movie Feactures')
plt.ylabel('Movie Feactures')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
